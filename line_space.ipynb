```python
def line(vec):
    xs = [i*vec[0] for i in range(-100, 100)]
    ys = [i*vec[1] for i in range(-100, 100)]
    zs = [i*vec[2] for i in range(-100, 100)]
    
    return xs, ys, zs
```

```python
import numpy as np
def space(vec1, vec2):
    vec1 = np.array(vec1)
    vec2 = np.array(vec2)
    x = np.linspace(-100, 100, 201)
    X = np.tile(x, (201, 1))
    Y = np.transpose(X)
    Z = np.zeros((201,201))
    for i in range(201):
        for j in range(201):
           A = np.array([[vec1[0],vec2[0]],[vec1[1],vec2[1]]])
           b = np.array([X[i][j],Y[i][j]])
           C = np.linalg.solve(A,b)
           Z[i][j] = np.dot(np.array([vec1[2],vec2[2]]),C)
    return X,Y,Z     
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

size = 150

xs, ys, zs = line([-1, -1, 1])

xss, yss, zss = space([1, 0, 1], [0, 1, 1])

fig = plt.figure(figsize = (5,5))
ax = fig.gca(projection = '3d')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")



ax.plot(xs, ys, zs, c= 'r')
ax.plot([-size, size], [0,0], [0,0], c= 'k')
ax.plot([0,0], [-size, size], [0,0], c= 'k')
ax.plot([0,0], [0,0], [-size, size], c= 'k')
ax.plot_surface(xss,yss,zss, alpha = 0.5)




ax.set_xlim(-size, size)
ax.set_ylim(-size, size)
ax.set_zlim(-size, size)


plt.show()

```
